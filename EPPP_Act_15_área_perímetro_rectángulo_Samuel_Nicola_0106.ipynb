📝 REDACCIÓN DEL PROBLEMA
En el estudio de la geometría, el rectángulo es una de las figuras más comunes y útiles. Un rectángulo es un cuadrilátero con cuatro ángulos rectos (90°). Para resolver problemas prácticos como calcular la cantidad de pintura necesaria para una pared o la cerca requerida para un terreno, necesitamos determinar dos medidas fundamentales:
### Formulas a utilizar
1. Area de rectangulo
* A = b * h
2. Perimetro del rectangulo
* P = 2 * ( b + h )

🎯 CARACTERÍSTICAS DEL PROGRAMA
* Estructura modular con 3 funciones independientes
* Menú interactivo con emojis y diseño elegante
* Uso de estructuras de control: if, elif, else, while
* Llamadas a funciones desde el menú principal
* Solicitud de base y altura dentro de cada función
* Resultados formateados con 2 decimales
* Sin programación orientada a objetos
* Código en un solo archivo listo para ejecutar

```python

def menu():
    print("\n╔══════════════════════════════════════════╗")
    print("║        🟦 MENÚ PRINCIPAL 🟦              ║")
    print("╠══════════════════════════════════════════╣")
    print("║  1. Calcular área del rectángulo    📐  ║")
    print("║  2. Calcular perímetro del rectángulo 📏 ")
    print("║  3. Salir del programa              👋  ║")
    print("╚══════════════════════════════════════════╝")
    return input("\n💡 Seleccione una opción (1-3): ")

def area_rectangulo():
    print("\n📐 CÁLCULO DEL ÁREA DEL RECTÁNGULO")
    print("─" * 40)
    base = float(input("Ingrese el valor de la base: "))
    altura = float(input("Ingrese el valor de la altura: "))
    area = base * altura
    print(f"\n✅ Resultados:")
    print(f"   Base: {base}")
    print(f"   Altura: {altura}")
    print(f"   Área: {area:.2f} unidades cuadradas")
    print("─" * 40)

def perimetro_rectangulo():
    print("\n📏 CÁLCULO DEL PERÍMETRO DEL RECTÁNGULO")
    print("─" * 40)
    base = float(input("Ingrese el valor de la base: "))
    altura = float(input("Ingrese el valor de la altura: "))
    perimetro = 2 * (base + altura)
    print(f"\n✅ Resultados:")
    print(f"   Base: {base}")
    print(f"   Altura: {altura}")
    print(f"   Perímetro: {perimetro:.2f} unidades")
    print("─" * 40)

print("\n" + "✨" * 40)
print("   BIENVENIDO AL SISTEMA DE CÁLCULO DE RECTÁNGULOS")
print("✨" * 40)

opcion = ""

while opcion != "3":
    opcion = menu()

    if opcion == "1":
        area_rectangulo()
    elif opcion == "2":
        perimetro_rectangulo()
    elif opcion == "3":
        print("\n" + "🎉" * 30)
        print("   ¡Gracias por usar el programa!")
        print("   👨‍🎓 Desarrollado para clase de Programación")
        print("🎉" * 30 + "\n")
    else:
        print("\n⚠️  Opción no válida. Por favor, seleccione 1, 2 o 3.")

    if opcion != "3":
        input("\n⏸️  Presione ENTER para continuar...")

print("✅ Programa finalizado exitosamente.\n")
print("Programa realizado por Samuel Nicola NL = 40")
```
